```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Amazon _Soft_toys_Scrping_Cleaned_data.csv")
```

## 1. Brand Performance Analysis

```python
df["Brand"].unique()
```

## Brand Frequency

```python
brand_counts = df["Brand"].value_counts()
brand_counts
```

```python
top5_brand = brand_counts.head()
top5_brand
```

## Average Rating by Brand

```python
brand_rating = df.groupby("Brand")["Rating"].mean().sort_values(ascending = False).round(2)
brand_rating
```

## Visualization: Top 5 brands by frequency

```python
plt.figure(figsize=(8,5))
sns.barplot(x=top5_brand.index,y=top5_brand.values,hue= top5_brand.index,palette="viridis")
plt.title("Top 5 Brands by Frequency")
plt.ylabel("Number of Products")
plt.xlabel("Brand")
plt.tight_layout()
plt.show()
```

## Visualization: Pie chart of top 5 brands

```python
plt.figure(figsize=(7,7))
plt.pie(top5_brand, labels=top5_brand.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('viridis', n_colors=5))
plt.title('Percentage Share of Top 5 Brands')
plt.tight_layout()
plt.show()
```

## Actionable Insights

```python
print("Top 5 Brands by Frequency:\n", top5_brand)
```

```python
print("Top 5 Brands by Average Rating:\n", brand_rating.head(5))
```

```python
print("High-rated but less frequent brands (potential gems):\n", brand_rating[~brand_rating.index.isin(top5_brand.index)].head(5))
```

## 2. Price vs. Rating Analysis

## Scatter Plot: Price vs. Rating

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x='Rating', y='Selling Price', hue='Brand', alpha=0.7)
plt.title('Price vs. Rating')
plt.xlabel('Rating')
plt.ylabel('Selling Price (INR)')
plt.legend(bbox_to_anchor=(1,1))
plt.tight_layout()
plt.show()
```

## Bar Chart: Average Price by Rating Range

```python
df['Rating Range'] = pd.cut(df['Rating'], bins=[0,3.9,4.2,4.5,5], labels=['<=3.9','4.0-4.2','4.3-4.5','4.6-5.0'])
avg_price_by_rating = df.groupby("Rating Range")["Selling Price"].mean()
plt.figure(figsize=(7,5))
sns.barplot(x=avg_price_by_rating.index, y=avg_price_by_rating.values,hue = ['<=3.9','4.0-4.2','4.3-4.5','4.6-5.0'], palette='magma')
plt.title('Average Price by Rating Range')
plt.xlabel('Rating Range')
plt.ylabel('Average Price (INR)')
plt.tight_layout()
plt.show()
```

## High value: high rating, low price

```python
high_value = df[(df['Rating'] >= 4.3) & (df['Selling Price'] < df['Selling Price'].median())]
print("\nHigh Value Products (High Rating, Low Price):")
print(high_value[['Title','Brand','Rating','Selling Price']].head(5))
```

## Overpriced but poorly rated

```python
overpriced_poorly_rated = df[(df['Rating'] < 4.0) & (df['Selling Price'] > df['Selling Price'].median())]
print("Overpriced but Poorly Rated Products:")
print(overpriced_poorly_rated[['Title','Brand','Rating','Selling Price']].head(5))
```

# 3. Review & Rating Distribution

## Top 5 Products by Reviews

```python
top5_reviews = df.sort_values('Reviews', ascending=False).head(5)
top5_reviews
```

## Top 5 Products by Rating

```python
top5_ratings = df[df['Reviews'] > 10].sort_values('Rating', ascending=False).head(5)
top5_ratings
```

## Bar Chart: Top-rated products

```python
plt.figure(figsize=(16, 8))
sns.barplot(x='Title', y='Rating', data=top5_ratings, hue='Title', palette='crest', legend=False)
plt.xticks(rotation=45, ha='right')
plt.title('Top 5 Highest Rated Products')
plt.ylabel('Rating')
plt.xlabel('Product Title')
plt.tight_layout()
plt.show()
```

## Bar Chart: Most-reviewed products

```python
plt.figure(figsize=(9,5))
sns.barplot(x='Title', y='Reviews',hue='Title', data=top5_reviews, palette='flare',legend=False)
plt.title('Top 5 Most Reviewed Products')
plt.ylabel('Number of Reviews')
plt.xlabel('Product Title')
plt.xticks(rotation=30, ha='right')
# plt.tight_layout()
plt.show()
```

## Actionable Insights

```python
print("\nTop 5 Most Reviewed Products:\n", top5_reviews[['Title','Brand','Reviews','Rating']])
```
